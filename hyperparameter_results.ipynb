```python
%matplotlib inline 
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
costs = pickle.load(open("hyperparam_search_costs/cost_outputs.p", "rb"))
```

```python
lr_a = []
mom_a = []
value_a = []
lr_m = []
mom_m = []
value_m = []

for key in costs.keys():
    if key[0] == "adam":
        lr_a.append(key[1])
        mom_a.append(key[2])
        try:
            value_a.append(np.mean(costs[key][-5:]))
        except:
            print costs[key]
            raise ValueError
    else:
        lr_m.append(key[1])
        mom_m.append(key[2])
        value_m.append(np.mean(costs[key][-5:]))
```

# hyperparameter results: adam optimzer

```python
plt.figure(figsize=(20,10))
aplot = plt.scatter(lr_a,mom_a,c = value_a, s=200)
plt.colorbar(aplot)
plt.title('adam optimzer color map of costs')
plt.xlabel('learning rate')
plt.ylabel('momentum')
plt.show(aplot)
```

```python
indexes= sorted(range(len(value_a)),key=lambda i: value_a[i])[:5]
print 'list of lowest costs, adam optimizer:'
print 'costs: ' + str([value_a[index] for index in indexes])
print 'learning rates: ' + str([lr_a[index] for index in indexes])
print 'momentums: ' + str([mom_a[index] for index in indexes])
```
